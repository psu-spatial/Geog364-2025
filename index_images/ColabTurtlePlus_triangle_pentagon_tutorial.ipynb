```python
# 2Ô∏è‚É£ Import the library
from ColabTurtlePlus.Turtle import *  # Screen, Turtle, helper functions
```

```python
# 3Ô∏è‚É£ Set up the drawing window
canvas = Screen()
canvas.setup(1000, 500)
canvas.bgcolor('aliceblue')
```

```python
# 4Ô∏è‚É£ Create and style your first turtle: Bob
clearscreen()
bob = Turtle()
bob.shape('turtle')
bob.shapesize(1.5)
bob.pencolor('blue')
bob.fillcolor('light blue')
bob.pensize(3)
bob.speed(2)
```

```python
# 5Ô∏è‚É£ Add a second turtle: Suri
Suri = Turtle()
Suri.shape('turtle2')
Suri.pencolor('#b22222')
Suri.fillcolor('#ffa07a')
Suri.pensize(2)
Suri.speed(8)
```

---

## 6Ô∏è‚É£ Make **Bob** draw an equilateral triangle
A triangle has 3 equal sides.  After each side Bob needs to turn **120¬∞** (because `360¬∞ / 3 = 120¬∞`).

```python
# Draw a triangle with side length 150 px
bob.penup()
bob.goto(-200, 0)   # move out of the way
bob.pendown()

for _ in range(3):
    bob.forward(150)
    bob.left(120)
```
